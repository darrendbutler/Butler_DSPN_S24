# Background
Provide a paragraph motivating the project. Why should someone outside your field care about this research?

Modern perceptions of distance learning or e-learning are dominated by internet-connected smart devices such as desktop computers, laptops, or smartphones. However, online learning creates infrastructure requirements that some learners may not be able to meet. For example, Nyemike Simeon et al found that for Nigerian and Ugandan students, the cost of internet services and internet inaccessibility were barriers to adopting e-learning during COVID-19 school closures (Nyemike Simeon et al., 2022).   Given that the digital divide in education is still a prevalent issue in both Global North and South countries, (Cheshmehzangi et al., 2023; Di Pietro, 2021; Gu, 2021; Martens et al., 2020; Nayak & Alam, 2022; Park et al., 2021), and access to education has a bearing on critical aspects of societal life, such as political action (Lindgren et al., 2019), people in low infrastructure environments would benefit a low infrastructure learning technologies to access educational content.

Interactive audio instruction has been used as a low infrastructure means of disseminating educational content (Moloo et al., 2023; United Nations Educational Scientific & Cultural Organization, 2021), historically as well as recently (e.g. in response to the COVID-19 pandemic).  Examples of audio instruction include interactive radio instruction, interactive voice response courses (Dione et al., 2021; Islam et al., 2022; Moloo et al., 2018), audio lectures (Bickerdike et al., 2014; White, 2009), and educational podcasts (Hew, 2009; Koenig, 2021; Lin et al., 2023; Rahmia et al., 2021).  

Interactive radio instruction (IRI), has been an effective low-infrastructure method for distance learning (Ho et al., 2009; Nwokedi et al., 2023; Raza, 2022; Verlumun Celestine et al., 2022). Tilson and Thomas defined interactive radio instruction as “radio broadcasted Instruction using interactive scripts written to solicit verbal or written student responses” (Tilson, 1991). Early interactive radio instruction practices cultivated the “interaction” component through supplemental material such as workbooks or the inclusion of caregivers and teachers in activities (Bosch & Crespo, 1995; Fryer, 1991; McBurnie, 2020).  However, recently interactive radio instruction projects have leveraged secondary communication technologies to aid in communication and assessment (STERLING, 2011; Yiya Solutions, 2022). The motivation for pairing interactive radio instruction is well stated by James Bontempo, a practitioner within the Information Communication Technologies for Development space (ICTD). Bontempo states, “Now that mobile phones are nearly ubiquitous…. why can't we use them to "close the loop" back to the instructor?... why not use ICT - mobile voice &/or SMS - to connect the learners directly back to the point of instruction" (Bontempo, 2009)? 

However, collecting data on student behavior can be challenging in educational radio programs. For example, in their 2021 handbook on interactive audio instruction, the United Nations Educational, Scientific and Cultural Organization (UNESCO), states a limitation of traditional radio as it is  “challenging to gather usage statistics and other monitoring and evaluation (M&E) data for radio” (United Nations Educational Scientific & Cultural Organization, 2021).  In framing secondary ICT devices as a means of scaling assessment and communication between instructors and learners, integrating secondary ICT devices is a major advancement within the IRI instructional method.  

The integration of secondary ICT devices is especially helpful for bridging the gap in gathering usage statistics, monitoring, and evaluation data for educational radio. Now, educators and education researchers can empirically measure student engagement in these programs. However, there is a lack of research investigating how different forms of student engagement may lead to different outcomes in students participating in interactive radio instruction. 

In order to gain insight into how different levels of student engagement may relate to student outcomes, we measured whether synchronous or asynchronous engagement matter in obtaining a certificate of completion. Syncrhonous engagement refers to the number of new practice questions a student attempts during the broadcast time (10:00-12:00), which may indicate if students are following along with broadcasts. Asynchronous engagement refers to the number of new practice questions a student attempts outside of broadcast time. 

We chose to investigate synchronous and asynchronous engagement as measures because they may relate to how cognitively engaged a student Without the ability to gauge usage, we cannot gauge student engagement, which has been found to explain differences in learning among students. The ICAP (Interactive, Constructive, Active, and Passive) Framework is a framework that links different levels of cognitive engagement to observable behaviors in students that lead to active learning outcomes. The ICAP framework suggests as students become more cognitively engaged with learning materials, from passive to active to constructive to interactive, their learning will increase (Chi & Wylie, 2014). If we were to map this framework onto student behavior in interactive radio instruction, a student who is listening to the radio without performing any activities may be less cognitively engaged than a student who is listening to the radio and answering questions throughout the broadcast. Additionally, a student who is answering assessment questions in response to a teacher reading the radio script may be more engaged than a student who is answering assessment questions without the radio teacher’s prompts. Figure 1 describes how cognitive engagement changes with student behavior in interactive radio instruction within the ICAP framework.

If synchronous engagement  is associated with a higher likelihood that a student receives a certificate, we hypothesize that a stronger relationship exists between a student’s synchronous engagement and the likelihood they receive a certificate performance than asynchronous participation and receiving a certificate. 

The purpose of this study is to better understand which student behaviors are associated with improved academic performance. By understanding which student behaviors are associated with certification, we may be able to design future audio instruction in a format that encourages students to adopt more beneficial behaviors. These findings may apply to other forms of audio instruction such as audio lectures or educational podcasts.

### Logstic Regression

```r
# Fit logistic regression model
glm.fit <- glm(rc ~ scaled_synch * scaled_asynch, data = learner_outcomes_df, family = binomial())

# Output the summary of the model to see results
summary(glm.fit)
```

If we accept that p must be less than 0.05, then it appears that both synchronous and asynchronous engagement predicts whether or not a student gets a certificate. More importantly, asynchronous engagement also seems to have a stronger effect on the likelihood of a student being classified as receiving a certificate. However, we should assess model fit by making predictions. 

```r
par(mfrow=c(2,2))
par(bg = "white")
plot(glm.fit)
```

There seems to be high leverage points impacting the model fit. To better assess model fit, we will use prediction. First we will look at the predicted probabilities of each student receiving a certificate (rc=1).

```r
# See how well the model captured the training set
glm_prob_df = data.frame(
    predicted_prob = predict(glm.fit, type = "response"),
    rc = learner_outcomes_df$rc)
#colnames(glm_prob_df) = c('predicted_prob')

num_observations = nrow(glm_prob_df)
glm_prob_df$index = seq(1, num_observations) #get the observation number for plotting; ggplot does not infer this

# Setting type="response" tells R to output prob. from P(Y=1,X) for all values of X
# "If no data set is supplied to the predict() function, then the probabilities 
# are computed for the training data that was used to fit the logistic regression model."

ggplot(glm_prob_df, aes(x= index, y=predicted_prob, color=rc)) + 
    geom_point() + 
    xlab('observation number') + 
    ylab('predicted response probability') 
```

```r
# First make a list of 0's for rc
glm_prob_df$predicted_binary <- rep(0, num_observations)

# Then use the probability output to label the certified students.
# Let's use a threshold of 50% probability.
threshold <- 0.50 # binarizing threshold
# find the rows that have prob > threshold and cast as 1
glm_prob_df$predicted_binary[glm_prob_df$predicted_prob > threshold] <- 1 

# Now let's look at the prediction accuracy
confusion_df <- data.frame(glm_prob_df$predicted_binary, learner_outcomes_df$rc)
colnames(confusion_df) <- c("predicted", "actual")

table(confusion_df)

# We can calculate the prediction accuracy by counting the number of times our
# prediction vector matched the real data and taking the average
print(paste("Accuracy:",mean(confusion_df$predicted == confusion_df$actual)))
print(paste("Error:",1 - mean(confusion_df$predicted == confusion_df$actual)))
```

The Logistic Regression gives an accuracy of 0.846. However, the model seems to be over-predicting non-certified students (rc=0). We can see that only 15/101 certified-students are being correctly  predicted.

However, this is error on training data so we need to. The actual performance of the model at predicting unseen values may be lower. 

## Cross Validation


### K-Fold Cross Validation
Let's try predicting on unseen data. We will use K=10 folds because these values have been shown to empiraclly better balance bias-variance tradeoff.

```r
library(boot) 
set.seed(1) #set the seed for the random number generator. resetting the seed will give different results.

# Set the cost function for binary classification
cost <- function(r, pi = 0) mean(abs(r-pi) > 0.5)
# Fit logistic regression model
glm.fit <- glm(rc ~ scaled_synch + scaled_asynch, data = learner_outcomes_df, family = binomial())
# See the cross validation error
cv.err = cv.glm(learner_outcomes_df, glm.fit, K=10)
print(cv.err$delta[1])
```

We have a small test Mean Squared Error of 0.097. This suggests that we have a good model fit.

### Bootstrapping
Bootstrapping quantifies uncertainty by reusing the data. We can boostrap our coefficients for synchronous and asynchronous engagement. 

```r
library(boot)

# The function needs two inputs: Data, Index
boot.fn <- function(data, index){  
    # return: throw this as output
    return(coef(glm(
        rc ~ scaled_synch + scaled_asynch, 
        data = learner_outcomes_df, family = binomial()))
    )
}
#test boostrap function
print(boot.fn(learner_outcomes_df, 1:nrow(learner_outcomes_df))) # note: "1:392" means all the rows
#Show coeficients from original glm
print(coef(glm.fit))
```

We have the same output from our earlier logistic regression so the boot function works as expected.

```r
boot_obj = boot(learner_outcomes_df, boot.fn ,R=1000) #R=repetitions 
print(boot_obj) #t1 is the intercept and t2 is the horsepower coeff.
```

The non parametric bootstrapped coefficients we receive for synchronous (t2) and asynchronous engagement are similar to the parametric estimates with standard errors close to zero. This suggests that there is little overlap in the coefficients. 

## Model Selection
We can compare the AIC criteria of models using only synch and asynch to determine which model performs better.

```r
glm.fit_synch <- glm(rc ~ scaled_synch, data = learner_outcomes_df, family = binomial())
glm.fit_asynch <- glm(rc ~ scaled_asynch, data = learner_outcomes_df, family = binomial())

summary(glm.fit_asynch)
summary(glm.fit_synch)
```

We see that the AIC for a model using only synchronous engagement is 521.69, the AIC for a model using only asynchronous engagement is 521.69. Because the asynchronous model has a lower AIC score, we can determine that asyncrhonous engagement better predicts the likelihood of receiving a certificate.

```r
# Load necessary libraries
library(rpart)
library(rpart.plot)

# Check the structure of the data frame to ensure the columns exist
str(learner_outcomes_df)

# Ensure 'rc' is a factor for classification
learner_outcomes_df$rc <- as.factor(learner_outcomes_df$rc)

# Train the decision tree model
decision_tree_model <- rpart(rc ~ scaled_asynch + scaled_synch, data = learner_outcomes_df, method = "class")

# Visualize the decision tree
rpart.plot(decision_tree_model, main = "Decision Tree for Predicting rc",
           type = 3, extra = 101, fallen.leaves = TRUE)

# Optional: Evaluate the model
# Split the data into training and testing sets (70% training, 30% testing)
set.seed(123) # For reproducibility
train_index <- sample(seq_len(nrow(learner_outcomes_df)), size = 0.7 * nrow(learner_outcomes_df))
train_data <- learner_outcomes_df[train_index, ]
test_data <- learner_outcomes_df[-train_index, ]

# Train the model on the training data
decision_tree_model <- rpart(rc ~ scaled_asynch + scaled_synch, data = train_data, method = "class")

# Predict on the test data
predictions <- predict(decision_tree_model, newdata = test_data, type = "class")

# Create a confusion matrix to evaluate the model
confusion_matrix <- table(Predicted = predictions, Actual = test_data$rc)
print(confusion_matrix)

# Calculate accuracy
accuracy <- sum(diag(confusion_matrix)) / sum(confusion_matrix)
print(paste("Accuracy:", accuracy))
```

## Create Pivot Tables

```r
create_pivot_tables <- function(df) {
  # Load necessary library
  library(dplyr)
  
  # Define the function to scale columns and add new columns
  scale_and_add_columns <- function(df) {
    df <- df %>%
      mutate(
        responses_within_broadcasts_zscored = scale(total_responses_during_broadcasts),
        responses_outside_broadcasts_zscored = scale(total_responses_outside_broadcasts)
      )
    return(df)
  }
  
  # Apply the function to the dataframe
  df <- scale_and_add_columns(df)
  
  # Print the resulting dataframe structure
  str(df)
  
  # Calculate the mean of received_certificate across categories in each categorical variable
  categorical_vars <- c('profile.access.radio', 'profile.age', 'profile.children', 
                        'profile.enrolled.in.school', 'profile.gender', 'profile.neighborhood.type', 
                        'profile.region.central', 'profile.region.eastern', 'profile.region.northern', 
                        'profile.region.western', 'profile.schooling', 'payment_type')
  
  # Create the directory if it does not exist
  dir.create("final_project/analysis_data_appendix", recursive = TRUE, showWarnings = FALSE)
  
  # Initialize an empty list to store pivot tables
  pivot_tables <- list()
  
  # Loop through each categorical variable to create pivot tables and save them as CSV
  for (var in categorical_vars) {
    pivot_table <- df %>%
      group_by(across(all_of(var))) %>%
      summarise(mean_received_certificate = mean(received_certificate, na.rm = TRUE)) %>%
      arrange(across(all_of(var)))
    
    # Define the file path for each pivot table
    file_path <- paste0("final_project/analysis_data_appendix/", var, "_pivot_table.csv")
    
    # Save the pivot table as a CSV file
    write.csv(pivot_table, file = file_path, row.names = FALSE)
    
    # Add the pivot table to the list with the file path as the key
    pivot_tables[[file_path]] <- pivot_table
  }
  
  # Print a message confirming the completion
  print("Pivot tables have been saved to final_project/analysis_data_appendix.")
  
  # Return the list of pivot tables
  return(pivot_tables)
}

# Call the function with learner_outcomes_full_df and store the result
pivot_tables <- create_pivot_tables(learner_outcomes_full_df)

# Print the keys of the pivot_tables list to confirm
print(names(pivot_tables))
```

### Plot Bar Plots

```r
create_bar_plots <- function(pivot_tables) {
  # Load necessary library
  library(ggplot2)
  
  # Initialize an empty dictionary to store bar plots
  bar_plots <- list()
  
  # Loop through each pivot table
  for (file_path in names(pivot_tables)) {
    # Extract the pivot table and file name
    pivot_table <- pivot_tables[[file_path]]
    file_name <- gsub(".csv", "", basename(file_path))
    
    # Create bar plot
    plot <- ggplot(pivot_table, aes(x = reorder(.data[[names(pivot_table)[1]]], -mean_received_certificate), y = mean_received_certificate)) +
      geom_bar(stat = "identity", fill = "skyblue", color = "black") +
      geom_text(aes(label = round(mean_received_certificate, 2)), vjust = -0.5, size = 3, color = "black") +
      labs(title = file_name, x = names(pivot_table)[1], y = "Mean Received Certificate") +
      theme_minimal() +
      theme(panel.background = element_rect(fill = "white", color = "black"),
            axis.text.x = element_text(angle = 45, hjust = 1))
    
    # Save the bar plot to file
    file_path_plot <- paste0("analysis_data_appendix/", file_name, "_bar_plot.png")
    ggsave(plot, file = file_path_plot, width = 8, height = 6, dpi = 300)
    
    # Add the bar plot to the dictionary with the file name as key
    bar_plots[[file_name]] <- plot
  }
  
  # Return the dictionary of bar plots
  return(bar_plots)
}

# Call the function with the pivot_tables list and store the result
bar_plots <- create_bar_plots(pivot_tables)

# Print the keys of the bar_plots dictionary to confirm
print(names(bar_plots))
```
